```python
x = np.linspace(-10, 10, 100)
x
```

```python
y = (np.sqrt(3 + x) - np.sqrt(3))/ x
y
```

```python

plt.plot(x, y)
```

```python
import math
sq = math.sqrt

def f(x):
    y = (sq(3 + x) - sq(3))/ x
    return y

for element in [0.1, 0.01, 0.001, 0.0001]:
    e = element
    a, b = f(-e), f(e)
    a, b = round(a, 4), round(b, 4)
    print(f"e : {e} \t f(-e) = {a}, f(e) = {b}")
```
