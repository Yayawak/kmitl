```python
import matplotlib.pyplot as plt

def estimateLimit(streval, n):
    arr = [10**-x for x in range(0, n)]
    
    data = {'hs': [],
           'y0s': [],
           'y1s': []
           }
    for h_ in arr:
    # +
        h = h_
        y0 = eval(streval)
    # -
        h *= -1
        y1 = eval(streval)
        
        print("left lim x->{} = {}".format(h, round(y1, 4)))
        print("right lim x->{} = {}".format(h, round(y0, 4)))
        print()
        
        data['hs'] += [h_]
        data['y0s'] += [y0]
        data['y1s'] += [y1]
    
    # plt.xscale("log")
    # plt.scatter(data['hs'], data['y0s'])
    # plt.scatter(data['hs'], data['y1s'])
    
    plt.plot(data['hs'], data['y0s'])
    plt.plot(data['hs'], data['y1s'])
    
    mn = min(data['y0s'] + data['y1s'])
    mx = max(data['y0s'] + data['y1s'])
    plt.vlines(data['hs'], mn, mx, 'green')
        
        
    print(tabulate(data, headers=['h', 'left', 'right']))
    
    # return data
strev = "(2.7**h - 1) / h"
data = estimateLimit(strev, 6)
data
```
