```python
import math
import random
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
r = 1.0
N = 1001
d = {"Trials": [], "Pi": []}
```

# We knows that it's uniform distribution 
### (Every darts can go to every point (pixel) equally)

```python
for Trial in range(1, N):
    circle_p = 0
    # important
    for i in range(Trial):
        # every position has possibility at the same value
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        
        if math.sqrt(x**2 + y**2) <= r:
            circle_p += 1
        
    d["Trials"].append(Trial)
    d["Pi"].append((circle_p / Trial) * 4)
```

```python
# k = N - 10
k = 10
# d['Trials'][:k], d['Pi'][:k]
# list(zip(d['Trials'][k:], d['Pi'][k:]))
list(zip(d['Trials'][:k], d['Pi'][:k]))
```

```python
df = pd.DataFrame(data=d)
df
```

```python
df = pd.DataFrame(data=d)
plt.figure(figsize = (10,7))
plot = sns.scatterplot(x="Trials", y="Pi", s=30, marker="o", data=df)
plot.set(title='Monte Carlo Simulation to Estimate Value of Pi',
         xlabel="Number of Trials", ylabel="Value of Pi")
plt.axhline(y=3.14, color='r', linestyle='-')
plt.show()
```
