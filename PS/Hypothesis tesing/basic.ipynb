# Z test

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np
```

```python
significance = 0.05 #alpha value
successes = np.array([320, 300])
samples = np.array([400, 425])
```

```python
stat,p_value = proportions_ztest(count=successes,nobs=samples,alternative='two-sided')
```

```python
print('z_stat: %0.5f, p_value: %0.5f' % (stat, p_value))
if p_value < significance:
    print ("Reject the null hypothesis")
else:
    print ("Accept the null hypothesis")
```

# T Test

```python
# T-test use to test sample
import numpy as np
import scipy.stats as stats
```

```python
significance = 0.05
A = np.array([43, 53, 65, 49, 55, 60, 47, 50, 60, 55])
B = np.array([62, 43, 54, 67, 59, 45, 46, 63, 65, 45])
```

```python
# NOT PAIR Test (independant samples)
stat, p_value = stats.ttest_ind(a = A, b = B, equal_var = True)
```

```python
print('t_stat: %0.5f, p_value: %0.4f' % (stat, p_value))
if p_value < significance:
    print ("Reject the null hypothesis")
else:
    print ("Accpet the null hypothesis")
```

```python
significance = 0.05
A = np.array([43, 53, 65, 49, 55, 60, 147, 50, 60, 55])
B = np.array([62, 43, 994, 67, 59, 45, 46, 63, 65, 45])
stat, p_value = stats.ttest_ind(a = A, b = B, equal_var = False)
print('t_stat: %0.5f, p_value: %0.4f' % (stat, p_value))
if p_value < significance:
    print ("Reject the null hypothesis")
else:
    print ("Accpet the null hypothesis")
```

### Paired t-Test :  is two samples are dependant ?
#### EX : Pre-test -> Post-test score 


```python
significance = 0.01
pre_test = np.array([60, 45, 80, 87, 79, 75, 60, 30, 45])
post_test = np.array([75, 65, 90, 80, 89, 95, 85, 69, 40])
stat, p_value = stats.ttest_rel(pre_test, post_test)
```

```python
print('t_stat: %0.5f, p_value: %0.4f' % (stat, p_value))
if p_value < significance:
    print ("Reject the null hypothesis")
else:
    print ("Accpet the null hypothesis")
```

# Chi Square Test
## Relationship between ***Categorical & Numeral samples***
- Marriage & Sex
#### If there is a relationship between gender and married status at  alpha =5%

   expected 
######  

```python
import pandas as pd
from scipy.stats import chi2_contingency
# contingency table
```

```python
df = pd.DataFrame(index=["Married","Single"],
                  data={'Male': [25, 35],
                        'Female': [15,25]})
df
```

```python
# c = 2 ; (sex)
# r = 2 ; (marriage status)
```

```python
chi2, p, dof, expected = chi2_contingency(df,
                                          #  2*2 
                                          #  n*n ; n > 2
    #   n < 5
                                          correction=False
                                         )
print(f"chi2 statistic: {chi2:.5g}")
print(f"p-value: {p:.5g}")
print(f"degrees of freedom: {dof}")
print("expected frequencies:")
print(expected)
```
