# Paired t-Test: Blood Pressure Difference


```python
import numpy as np
import pandas as pd
import scipy.stats as stats
df = pd.read_csv('BloodPressure.csv')
```

```python
df.head()
```

```python
df[['bp_before','bp_after']].describe()
```


## Dependant samples

```python
significance = 0.01
stat,p_value = stats.ttest_rel(df['bp_before'], df['bp_after'])
```

```python
print('t_stat: %0.5f, p_value: %0.4f' % (stat, p_value))
if p_value < significance:
    print ("Reject the null hypothesis")
else:
    # Thres is a statically significant decrease in blood pressure
    print ("Accpet the null hypothesis")
```
