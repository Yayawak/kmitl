```julia
function my_cumsum(x)
    y = similar(x)
    # allocate an array 
    y[1] = x[1]
    for i = 2:length(x)
        y[i] = y[i - 1] + x[i]
    end
    return y
end

my_cumsum
```

```julia
x = rand(Float32, 10 ^ 7) # [0, 1]
```

```julia
@time y = my_cumsum(x)
```

```julia
yexact = my_cumsum(Float64.(x)) # double presition
```

```julia
1 .- 8
```

```julia
1 - 8
```

```julia
err = abs.(y .- yexact) ./ abs.(yexact)
```

```julia
using PyPlot
```

```julia
n = 1:10:length(err)
```

```julia
n
```

```julia
loglog(n, err[n])

ylabel("relativec error")
xlabel("# summands")

loglog([1, length(err)], 
    sqrt.([1, length(err)]) * 1e-7,
    "k--"
)
text(1e3, 1e-5, L"\sim \sqrt{n}")
title("naive cumsum implementation")
```
